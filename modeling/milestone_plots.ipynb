```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ccy_list = ['nzdusd', 'usdchf', 'gbpusd', 'usdnok', 'usdsek', 'audusd', 'eurusd', 'usdcad', 'usdjpy']
```

```python
def plot(ccy, rewards):
    cmap = ['red', 'blue']

    # Plot L of class 0
    fig, ax = plt.subplots(1, 1, figsize=(10, 7.5))
    
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)

    # Provide tick lines across the plot to help your viewers trace along the axis ticks. 
    plt.grid(True, which='major', axis='y', ls='--', lw=0.5, c='k', alpha=0.3)
    
#    plt.tick_params(axis='both', which='both', bottom=True, top=True,
#                   labelbottom=True, left=True, right=True, labelleft=True)

    plt.xlabel(r"Episodes")
    plt.ylabel(r"Cumulative rewards")
    
    runs = ['Baseline 1 (DRL)', 'Baseline 2 (Traditional)']

    plt.title('Test results ('+ccy.upper()+')', fontsize=16)
    for i in range(2):
        plt.plot(range(1,len(rewards[0])+1), rewards[i], lw=2.5, color = cmap[i], label=runs[i])

    plt.legend(ncol=2)
    
    fig.savefig(ccy+'.png', figsize=(12,8))
    plt.show()
    fig.clear() 
```

```python
drl_rewards = {}
mr_rewards ={}
test_rewards = []

for ccy in ccy_list:
    drl_rewards[ccy] = pd.read_csv('../results/milestone/test_drl_'+ccy+'.csv')
    mr_rewards[ccy] = pd.read_csv('../results/milestone/test_mean_rev_'+ccy+'.csv')
    test_rewards.append(drl_rewards[ccy])
    test_rewards.append(mr_rewards[ccy])
    plot(ccy, test_rewards)
    test_rewards = []
```
